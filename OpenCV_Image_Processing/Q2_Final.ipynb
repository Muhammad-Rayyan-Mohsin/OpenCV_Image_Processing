```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
#-------------------------------------------------------------------------------------------------------------
# Function to rotate an image by a specified angle
def ImageRotation(image, angle):
    # Get the center of the image
    center = (image.shape[1] // 2, image.shape[0] // 2)

    # Perform the rotation
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated_image = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

    return rotated_image

# Function to resize an image to a specified width and height
def Img_Resizing(image, width, height):
    return cv2.resize(image, (width, height))

#---------------------------------------------------------------------------------------------------------------------

background_height = 3 * 346  # Double the height to accommodate two rows
background_width = 346 * 10  # Updated to accommodate additional sub-images
background = np.ones((background_height, background_width, 3), np.uint8) * 255

# Load your sub-images
Car1 = cv2.imread('/content/img1.png')
Car3 = cv2.imread('/content/img3.png')
Car5 = cv2.imread('/content/img5.png')
Car6 = cv2.imread('/content/img6.png')
Car7 = cv2.imread('/content/img7.png')
Car18 = cv2.imread('/content/img18.png')
Clock4 = cv2.imread('/content/img4.png')
Clock9 = cv2.imread('/content/img9.png')
Clock2 = cv2.imread('/content/img2.png')
Clock14 = cv2.imread('/content/img14.png')
Clock8 = cv2.imread('/content/img8.png')
Clock17 = cv2.imread('/content/img17.png')
Clock11 = cv2.imread('/content/img11.png')
Clock13 = cv2.imread('/content/img13.png')
Clock15 = cv2.imread('/content/img15.png')
Clock10 = cv2.imread('/content/img10.png')
Clock12 = cv2.imread('/content/img12.png')
Clock16 = cv2.imread('/content/img16.png')

#--------------------------------------------------------------------------------------------------------------------------------------------

# Resize all sub-images to 346x346 pixels
commonW, commonH = 346, 346
Car1 = Img_Resizing(Car1, commonW, commonH)
Car3 = Img_Resizing(Car3, commonW, commonH)
Car5 = Img_Resizing(Car5, commonW, commonH)
Car6 = Img_Resizing(Car6, commonW, commonH)
Car7 = Img_Resizing(Car7, commonW, commonH)
Car18 = Img_Resizing(Car18, commonW, commonH)
Clock4 = Img_Resizing(Clock4, commonW, commonH)
Clock9 = Img_Resizing(Clock9, commonW, commonH)
Clock2 = Img_Resizing(Clock2, commonW, commonH)
Clock14 = Img_Resizing(Clock14, commonW, commonH)
Clock8 = Img_Resizing(Clock8, commonW, commonH)
Clock17 = Img_Resizing(Clock17, commonW, commonH)
Clock11 = Img_Resizing(Clock11, commonW, commonH)
Clock13 = Img_Resizing(Clock13, commonW, commonH)
Clock15 = Img_Resizing(Clock15, commonW, commonH)
Clock10 = Img_Resizing(Clock10, commonW, commonH)
Clock12 = Img_Resizing(Clock12, commonW, commonH)
Clock16 = Img_Resizing(Clock16, commonW, commonH)

#-------------------------------------------------------------------------------------------------------------------

#img7
Car7_flip=cv2.flip(Car7,0)
Car7_rotated= cv2.rotate(Car7_flip, cv2.ROTATE_90_CLOCKWISE)
# Rotate img5 by 90 degrees to the right
Car5_rotated = ImageRotation(Car5, -90)

# Rotate img6 by 45 degrees
Car6_flip= cv2.flip(Car6,0)
Car6_rotated= cv2.rotate(Car6_flip, cv2.ROTATE_90_CLOCKWISE)


# Rotate img3 by 90 degrees to the right
Car3_rotated = ImageRotation(Car3, 45)

# Define positions for each sub-image
positions = [
    ((0, 0), Car1),
    ((0, commonW), Car5_rotated),
    ((0, commonW * 2), Car18),
    ((commonH, 0), Car7_rotated),
    ((commonH, commonW), Car3_rotated),
    ((commonH, commonW * 2), Car6_rotated)
]

for (row, col), sub_image in positions:
    if sub_image is not None:
        x1 = col
        x2 = col + sub_image.shape[1]
        y1 = row
        y2 = row + sub_image.shape[0]
        background[y1:y2, x1:x2] = sub_image

Clock4_rotated = ImageRotation(Clock4, -70)
Clock11_flip=cv2.flip(Clock11,1)
Clock9_flip=cv2.flip(Clock9,1)
Clock17_rotated = ImageRotation(Clock17, 90)
Clock13_rotated = ImageRotation(Clock13, -45)

# Define positions for additional sub-images
additional_positions = [
    ((0, 3 * commonW), Clock4_rotated),
    ((commonH, 3 * commonW), Clock9_flip),
    ((2 * commonH, 3 * commonW), Clock2),
    ((0, 4 * commonW), Clock14),
    ((commonH, 4 * commonW), Clock8),
    ((2 * commonH, 4 * commonW), Clock17_rotated),
    ((0, 5 * commonW), Clock11_flip),
    ((commonH, 5 * commonW), Clock13_rotated),
    ((2 * commonH, 5 * commonW), Clock12),
    ((0, 6 * commonW), Clock16),
    ((commonH, 6 * commonW), Clock10),
    ((2 * commonH, 6 * commonW), Clock15)
]

# Place each additional sub-image onto the background at the specified positions
for (row, col), sub_image in additional_positions:
    if sub_image is not None:
        x1 = col
        x2 = col + sub_image.shape[1]
        y1 = row
        y2 = row + sub_image.shape[0]
        background[y1:y2, x1:x2] = sub_image

#-------------------------------------------------------------------------------------------------------------------

# Save or display the complete image
cv2.imwrite('/content/complete_image.jpg', background)
cv2_imshow(background)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
